# **Scatter Plot**
直接用列名进行plot


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(123)

df = pd.DataFrame({'A':np.random.randn(365).cumsum(0),
           'B':np.random.randn(365).cumsum(0)+20,
           'C':np.random.randn(365).cumsum(0)-20},index=pd.date_range('2019-1-1', periods=365))
```

```
# 'A':X轴坐标
# 'B'：Y轴坐标
df.plot('A','B', kind='scatter');
```

```
# 加上colormap 相当于显示三维图像
df.plot.scatter('A', 'C', c='B', colormap='viridis', sharex=False);
```

```
# 加上大小，可以显示第四维度信息
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis');
```

```
# 按比例画出来之，可以看出A的范围要小于C
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal')
```

# **Box Plot**

```
df.plot.box()
```

# **Histogram**

```
df.plot.hist(alpha=0.7)
```

# **Density Plot**

```
df.plot.kde();
```
