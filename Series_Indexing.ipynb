## **一、中括号**
官方不推荐，容易有歧义<br>
.ix方法与中括号用法类似，不举例说明了


```
#情况一：当index为integer时，中括号代表index
import pandas as pd
george = pd.Series([10, 7], index=[1968, 1969], name='George Songs')
george[1968]
#george[0]  #会出错
```

```
#情况二：当index为非integer且index 为unique时，中括号既可以表示key，也可以表示位置
george = pd.Series([10, 7], index=['1968', '1969'], name='George Songs')
george
```

```
print(george['1968'])
print(george[0])
```

```
george = pd.Series([10, 7, 11], index=['1968', '1969', '1969'], name='George Songs')
#index不为unique，则不成立
george[0]
```

## iloc

```
import pandas as pd
g = pd.Series([10, 7, 1, 22], index=['1968', '1969', '1970', '1970'], name='George Songs')
g
```

```
g.iloc[0]
```

```
g.iloc[01]
```

```
#slice
g.iloc[0:3]
```

```
#list
g.iloc[[0,3]]
```

## loc

```
import pandas as pd
g = pd.Series([10, 7, 1, 22, 33, 23], index=['1968', '1969', '1970', '1970','1971', '1975'], name='George Songs')
g
```

```
g['1968']
```

```
#list
g[['1968', '1970']]
```

```
# slice:
g['1968':'1971']
```

# **四、Boolean Array**

```
g = pd.Series([10, 7, 1, 22], index=['1968', '1969', '1970', '1970'], name='George Songs')
g
```

```
# boolean array
mask = g > 7
mask
```

```
# 把boolean array带回去
g[mask]
```

```
mask2 = g < 3
mask2
```

```
g[mask | mask2]
```

```
g[(g > 7) | (g < 3)]
```
